```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.pandas.set_option('display.max_columns', None)

df = pd.read_csv('data/Chennai.csv')
df.head()
```

```python
sns.distplot(df['Area'], hist = False)
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot(df['Area'], df['No. of Bedrooms'], df['Price'], data=df, kind="box")
    g.set_axis_labels("Area", "No. of Bedrooms");
```
