```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
pd.pandas.set_option('display.max_columns', None)

#creating dataframe
df = pd.read_csv('data/mumbai.csv')
df.head()
```

```python
#observing co-realtion between labels
corrMatrix = df.corr()
corrMatrix
```

```python
#number are tough to interpret reasult:/ Lets try gradient!! HEATMAP
fig , ax = plt.subplots(figsize = (30,30))
sn.heatmap(corrMatrix, annot = True)
plt.show()
```

```python
##AREA and NO OF BEDROOM shows +ve realtion to price! 
#Lets drop others
```

```python
#plot price vs Area

x_area = np.array(df['Area'])
ys = np.array(df['Price'])

ax.bar(x_area, ys, width = 0.4)
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
#plot price vs No. of Bedrooms 
x_bedrooms = df['No. of Bedrooms']

plt.bar(x_bedrooms, ys, width = 0.4)
plt.xlabel('No. of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
X = np.arange(4)
ax.bar(X+0.00, x_area, color = 'r', width = 0.25)
ax.bar(X+0.50, x_bedrooms, color = 'g', width = 0.25)
```
